##1. Select: Subset Rows, Columns by Slicing

There are three principal methods for slicing rows and columns in a dataframe or series.

* <b>.loc</b> works on <b><i>labels</i></b> in the index
* <b>.iloc</b> works on <b><i>the positions</i></b> in the index (so it only takes integers)
* <b>.ix</b> usually tries to behave like loc but falls back to behaving like iloc if the label is not in the index

```
df
```

```
df.iloc[1:2, 1:2]
```

```
df.loc['Monaco':'United Kingdom', 'population':'gdp']
```

```
df.ix[1:, 'area':'gdp']
```

```
# we can also explicity subset columns by list
df1 = df[['area','population']]
```
