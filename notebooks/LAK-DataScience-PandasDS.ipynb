##1. Preliminaries

```
import numpy as np  
import pandas as pd
import matplotlib.pyplot as plt  
import seaborn as sns 
```

```
# We will create a datasets from these lists

countries = ['Monaco','Bangladesh', 'Tuvalu', 'United Kingdom','Singapore']

populations = [38400,162076426, 11323, 65110000, 5607300]

areas = [2.02, 143998.0,26,242910.0,719.2]

continents = ['Europe','Asia','Australia','Europe','Asia']

gdps = [7060,173062,32,2858003,307872]

labels = ['population','area','gdp','continents']
```

## 2. Pandas Series Object

```
# create series object
my_series = pd.Series(populations)
```

```
# a series is an indexed one dimensional array
my_series
```

```
# a variety of methods are available 
my_series.describe()
#my_series.mean()
#my_series.std()
```

```
# the index can be labels instead of default integers
my_series = pd.Series(populations, index=countries)
```

```
my_series
```

```
# we can think of the index label as a dictionary key; supplying key returns value
my_series.Monaco
```

```
# note the following is equivalent
my_series['Monaco']
```

```
# we can perform pythonesque slicing on rows
my_series[:2]
```

```
# we can also use index labels for slicing
my_series['Bangladesh':'United Kingdom']
```

## 3. Pandas DataFrame Object

A dataframe is a tabular data structure (rows, columns). The rows and columns don't have to be labeled.

```
df = pd.DataFrame(list(zip(populations,areas,gdps,continents)))
```

```
df
```

```
df.shape
```

```
df.index
```

```
df.columns
```

```
# let's recreate dataframe, this time with labels for both row and columns
# row labels are called indices; column labels are called columns
df = pd.DataFrame(list(zip(populations,areas,gdps,continents)), index=countries, columns=labels)
```

```
df
```

```
df.info()
```

```
# derive the mean of the area
df.area.mean()
```

```
# unique values in a column for a categorical variable
df.continents.unique()
```

```
# value counts for a categorical variable
df.continents.value_counts()
```
