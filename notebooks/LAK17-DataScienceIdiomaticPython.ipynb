### Looping over a collection

```
colors = ['red', 'green', 'blue', 'yellow']
```

```
for i in range(len(colors)):
  print(colors[i])
```

```
for color in colors:
  print(color)
```

### Looping backwards

```
for color in reversed(colors):
  print(color)
```

### Looping over collection and indices

```
for i in range(len(colors)):
  print(i, '--->', colors[i])
```

```
for i, color in enumerate(colors):
    print(i, '-->', color)
```

### Looping over two collections

```
mathematicians = ["Gauss", "Euler", "Ramanujan", "Turing", "Galois"]
countries = ["Germany", "Switzerland", "India", "England", "France"]
```

```
for mathematician, country in zip(mathematicians,countries):
  print(mathematician, '--->', country)
```

```
for mathematician in sorted(mathematicians):
  print(mathematician)
```

### Custom Sort Order

```
print(sorted(mathematicians, key=len))
```

## Dictionary Skills

###Looping over dictionary keys

```
d = {'Albania': 'Tirana', 'Burkina Faso': 'Ouagadougou', 'Nigeria': 'Abuja', 'San Marino':'San Marino', 'Taiwan':'Taipei'}
```

```
for k in d:
  print(k)
```

```
for k,v in d.iteritems():
  print(k,'-->',v)
```
